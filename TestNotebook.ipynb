```python
import boto3
ec2_resource = boto3.resource('ec2')
```

```python
help(ec2_resource.instances.filter)
```

```python
instances = ec2_resource.instances.filter(
    Filters=[
        {
            'Name': 'instance-state-name',
            'Values': ['running']
        },
        {
            'Name': 'tag:Project',
            'Values': ['SparkCluster']
        },
        {
            'Name': 'tag:User',
            'Values': ['root']
        },
        {
            'Name': 'tag:NodeType',
            'Values': ['Master']
        }
    ]
)
if list(instances):
    for instance in instances:
        print(instance)
else:
    print("No master node present for current user.")
```
